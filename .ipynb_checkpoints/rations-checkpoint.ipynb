```python
coeffs = []
with open("coeffs.txt", "r", encoding = "ISO-8859-1") as f:
    for line in f.readlines():
        line = line.strip().split('__')
        coeffs.append((line[0], line[1]))
coeffs
```

```python
from datetime import datetime
import time

def convertDatetoSeconds(date):
    try:
        d = datetime.strptime(date, "%H:%M:%S %d.%m.%Y")
        return time.mktime(d.timetuple())
    except:
        return None

updated_coeffs = []
for coeff in coeffs:
    conv = convertDatetoSeconds(coeff[1])
    if conv:
        updated_coeffs.append((coeff[0], conv))
len(updated_coeffs)
```

```python
import torch
from torch.utils.data import Dataset
from torch.utils.data import DataLoader
import numpy as np

class RatioDataset(Dataset):
    
    def __init__(self, ratios, seq_len = 100):
        self.seq_len = seq_len
        
        timestamps = []
        for i in range(0, len(ratios) - seq_len):
            timestamps.append(ratios[i:i+seq_len])
        self.timestamps = timestamps
    
    def __len__(self):
        return len(self.timestamps)
    
    def __getitem__(self, idx):
        timestp = self.timestamps[idx]
        times = [int(t[1]) for t in timestp]
        labels = [float(t[0]) for t in timestp]
        
        return np.array(times), np.array(labels)
    
dataset = RatioDataset(updated_coeffs)
loader = DataLoader(dataset, batch_size=4)

x, y = next(iter(loader))
y
```
